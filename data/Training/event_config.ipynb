```python
import json
import pandas as pd 

with open("event_config.json", "r") as f:
    e= json.load(f)
# Sample training data (replace this with your actual DataFrame)
training_data = pd.read_csv('TrainingData2.csv')
# Generate configuration based on the training data
event_config = {}
variable_dates_events = ['corona','ramadan_season', 'eid_al_fitr', 'eid_al_adha',
    'international_expo', 'oil_price_surge', 'midterm_break', 'air_show', 
    'expo_2020', 'hajj_season', 'royal_wedding', 'international_day_of_happiness', 
    'dubai_marathon', 'world_cup', 'economic_slowdown', 'shopping_festival', 
    'comicon_dubai', 'global_financial_tension', 'pre_hajj_season', 'formula_1_grand_prix','diwali'
] 
fixed_dates_events = [ 'new_year','uae_national_day', 'national_day_rush',
    'christmas_season', 'new_years_eve', 
    'dubai_shopping_festival', 'spring_break', 'winter_holiday_season', 'dubai_new_year','summer_season','spring_season','new_year_break'
]

for event in training_data['event'].unique():
    event_data = training_data[training_data['event'] == event]
    years = event_data['start_date'].apply(lambda x: pd.to_datetime(x).year).unique().tolist()
    avg_arrival = event_data['arrival_percentage_change'].tolist()
    avg_departure = event_data['departure_percentage_change'].tolist()
    start_dates = event_data['start_date'].unique().tolist()
    end_dates =event_data['end_date'].unique().tolist()

    event_config[event] = {
        "type":"",
        "source": e[event]['source'] if event in e.keys() else "" ,
        "years": years,
        "percentage_changes": {
            "arrival": avg_arrival,
            "departure": avg_departure
        },
        "dates": {"start":start_dates,
                  "end":end_dates}}
    if event in variable_dates_events:
        event_config[event]['type']='variable_dates_events'
    if event in fixed_dates_events:
        event_config[event]['type']='fixed_dates_events'
        event_config[event]['source']='fixed_event'


# Save configuration to JSON
with open("event_config1.json", "w") as f:
    json.dump(event_config, f, indent=4)

# Print the configuration
print(json.dumps(event_config, indent=4))
```
